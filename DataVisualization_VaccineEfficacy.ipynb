```
### Importing Required Libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```
### Reading the Data

df = pd.read_csv("/content/MainDataset(Vaccine).csv")
df.head()
```

```
### Data Visulization of Vaccine Efficacy

covid_positive = pd.value_counts(df['Covid Positive'])
names = ['Efficacious', 'Inefficacious']
plt.pie(covid_positive, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
plt.title('VACCINE EFFICACY', size = 25)
plt.show()
```

```
### Vaccine Efficacy based on Gender

covid_gender = df.iloc[:, [5,23]]
gender = covid_gender.groupby('Female')

male = gender.get_group(0)
female = gender.get_group(1)

covid_male = pd.value_counts(male['Covid Positive'])
covid_female = pd.value_counts(female['Covid Positive'])

names = ['Efficacious', 'Inefficacious']
fig = plt.figure()

ax1 = fig.add_axes([0, 1.5, 0.5, 0.5], aspect=1)
ax1.pie(covid_male, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax1.set_title('Male')
ax2 = fig.add_axes([0.5, 1.5, 0.5, 0.5], aspect=1)
ax2.pie(covid_female, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax2.set_title('Female')
plt.suptitle('Vaccine Efficacy based on Gender', x='1.05',y='2.2', ha ='right', va='top', size='25')
plt.show()
```

```
### Vaccine Efficacy amongst different Age Groups

covid_age18 = df.iloc[:, [6,23]]
age_18 = covid_age18.groupby('Age 0-18')
covid_age35 = df.iloc[:, [7,23]]
age_35 = covid_age35.groupby('Age 19-35')
covid_age50 = df.iloc[:, [8,23]]
age_50 = covid_age50.groupby('Age 36-50')
covid_age60 = df.iloc[:, [9,23]]
age_60 = covid_age60.groupby('Age 51-60')
covid_age_old = df.iloc[:, [10,23]]
age_old = covid_age_old.groupby('Age 60 Above')

_18 = age_18.get_group(1)
_35 = age_35.get_group(1)
_50 = age_50.get_group(1)
_60 = age_60.get_group(1)
_old = age_old.get_group(1)

covid_18 = pd.value_counts(_18['Covid Positive'])
covid_35 = pd.value_counts(_35['Covid Positive'])
covid_50 = pd.value_counts(_50['Covid Positive'])
covid_60 = pd.value_counts(_60['Covid Positive'])
covid_old = pd.value_counts(_old['Covid Positive'])

names = ['Efficacious', 'Inefficacious']
fig = plt.figure()

ax1 = fig.add_axes([0, 1.5, 0.5, 0.6], aspect=1)
ax1.pie(covid_18, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax1.set_title('0-18')
ax2 = fig.add_axes([0.5, 1.5, 0.5, 0.6], aspect=1)
ax2.pie(covid_35, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax2.set_title('19-35')
ax3 = fig.add_axes([1, 1.5, 0.5, 0.6], aspect=1)
ax3.pie(covid_50, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax3.set_title('35-50')
ax4 = fig.add_axes([0, 0.8, 1, 0.6], aspect=1)
ax4.pie(covid_60, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax4.set_title('51-60')
ax5 = fig.add_axes([0.5, 0.8, 1, 0.6], aspect=1)
ax5.pie(covid_old, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax5.set_title('60 Above')
plt.suptitle('Vaccine Efficacy based on Age Group', x='1.3',y='2.3', ha ='right', va='top', size='25')
plt.show()
```

```
### Vaccine Efficacy amongst Obese 

covid_obese = df.iloc[:, [11,23]]
obese = covid_obese.groupby('Obese Yes')

no_obese = obese.get_group(0)
yes_obese = obese.get_group(1)

covid_no_obese = pd.value_counts(no_obese['Covid Positive'])
covid_yes_obese = pd.value_counts(yes_obese['Covid Positive'])

names = ['Efficacious', 'Inefficacious']
fig = plt.figure()

ax1 = fig.add_axes([0, 1.5, 0.5, 0.5], aspect=1)
ax1.pie(covid_no_obese, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax1.set_title('Not Obese')
ax2 = fig.add_axes([0.5, 1.5, 0.5, 0.5], aspect=1)
ax2.pie(covid_yes_obese, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax2.set_title('Obese')
plt.suptitle('Vaccine Efficacy amongst Obese People', x='1.15',y='2.2', ha ='right', va='top', size='25')
plt.show()
```

```
### Vaccine Efficacy for people with various side effects (after vaccination)

muscle_pain = df.iloc[:, [17,23]]
pain_side_effect = muscle_pain.groupby('Muscle/ Joint pain')
tiredness = df.iloc[:, [18, 23]]
tired_side_effect = tiredness.groupby('Tiredness')
headache = df.iloc[:, [19,23]]
headache_side_effect = headache.groupby('Headache')
none = df.iloc[:, [20,23]]
none_side_effect = none.groupby('None')
fever = df.iloc[:, [21,23]]
fever_side_effect = fever.groupby('Fever')
mild_allergy = df.iloc[:, [22,23]]
allergy_side_effect = mild_allergy.groupby('Mild Allergies')

_muscle_pain = pain_side_effect.get_group(1)
_tired = tired_side_effect.get_group(1)
_headache = headache_side_effect.get_group(1)
_none = none_side_effect.get_group(1)
_fever = fever_side_effect.get_group(1)
_allergy = allergy_side_effect.get_group(1)

covid_muscle_pain = pd.value_counts(_muscle_pain['Covid Positive'])
covid_tired = pd.value_counts(_tired['Covid Positive'])
covid_headache = pd.value_counts(_headache['Covid Positive'])
covid_none = pd.value_counts(_none['Covid Positive'])
covid_fever = pd.value_counts(_fever['Covid Positive'])
covid_allergy = pd.value_counts(_allergy['Covid Positive'])

names = ['Efficacious', 'Inefficacious']
fig = plt.figure()

ax1 = fig.add_axes([0, 1.5, 0.5, 0.5], aspect=1)
ax1.pie(covid_muscle_pain, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax1.set_title('Muscle/Joint Pain')
ax2 = fig.add_axes([0.5, 1.5, 0.5, 0.5], aspect=1)
ax2.pie(covid_tired, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax2.set_title('Tiredness')
ax3 = fig.add_axes([1, 1.5, 0.5, 0.5], aspect=1)
ax3.pie(covid_headache, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax3.set_title('Headache')
ax4 = fig.add_axes([0, 0.9, 0.5, 0.5], aspect=1)
ax4.pie(covid_none, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax4.set_title('None')
ax5 = fig.add_axes([0.5, 0.9, .5, 0.5], aspect=1)
ax5.pie(covid_fever, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax5.set_title('Fever')
ax6 = fig.add_axes([1, 0.9, .5, 0.5], aspect=1)
ax6.pie(covid_allergy, labels=names, autopct="%0.2f%%", 
        explode=(0,0), shadow=True)
ax6.set_title('Mild Allergies')

plt.suptitle('Vaccine Efficacy amongst People with Side Effects', x='1.5',y='2.2', ha ='right', va='top', size='25')
plt.show()
```
